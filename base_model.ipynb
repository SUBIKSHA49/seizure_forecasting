```python
df = pd.read_csv("C:\Users\subim\OneDrive\Desktop\dl_cat2\Epileptic Seizure Recognition Data Set.csv")  # change path if needed

# Drop unnecessary column
df = df.drop(columns=['Unnamed'])

print(df.shape)
df.head()
```

```python
# Select one sample (for base model)
signal = df.iloc[0, :-1].values  # exclude label

# Convert to pandas series
time_series = pd.Series(signal)

plt.plot(time_series)
plt.title("EEG Signal Time Series")
plt.show()
```

```python
train_size = int(len(time_series) * 0.8)
train, test = time_series[:train_size], time_series[train_size:]
```

```python
result = adfuller(train)

print("ADF Statistic:", result[0])
print("p-value:", result[1])
```

```python
train_diff = train.diff().dropna()
```

```python
model = ARIMA(train, order=(2,1,2))
model_fit = model.fit()

print(model_fit.summary())
```

```python
forecast = model_fit.forecast(steps=len(test))

# Convert to array
forecast = np.array(forecast)
```

```python
mse = mean_squared_error(test, forecast)
rmse = np.sqrt(mse)
mae = mean_absolute_error(test, forecast)

print("MSE:", mse)
print("RMSE:", rmse)
print("MAE:", mae)
```

```python
plt.figure(figsize=(10,5))
plt.plot(test.values, label='Actual')
plt.plot(forecast, label='Forecast')
plt.legend()
plt.title("ARIMA Forecast vs Actual")
plt.show()
```
